# **Titanic Crisis Analysis — TCA**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.info()
```

## df["Age"] --> float64 which is actually int64

```python
df["Age"].values
```

```python
df["Age"] = df["Age"].fillna(
    df.groupby(["Sex", "Pclass"])["Age"].transform("median")
)
```

```python
df["Age"] = df["Age"].round(0).astype("Int64")
```

```python
df["Age"].values
```

```python
df["FamSize"] = df["SibSp"] + df["Parch"] + 1
df["FamSize"]
df
```

```python
df = df.drop(columns=["SibSp", "Parch"])
df
```

```python
col = df.pop("FamSize")
df.insert(6,"FamSize",col)
df
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.shape
```

# Name Extraction

```python
df["Title"] = df["Name"].str.extract(r", (\w+)\.")
df.groupby("Title")["Survived"].agg(["count", "mean"])
```
